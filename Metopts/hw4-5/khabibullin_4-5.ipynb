```python
import matplotlib.pyplot as plt
import numpy as np
from main_search import *
```

```python
func = [lambda x: 1/6*x**6 - 52/25*x**5 + 39/80*x**4+71/10*x**3-79/20*x**2-x+1/10,
       lambda x: np.sin(x)+np.sin(10/3*x),
       lambda x: -sum([k*np.sin((k+1)*x +k) for k in range(1, 6)]),
       lambda x: -(16*x**2-24*x+5)*np.exp(-x),
       lambda x: (3*x-1.4)*np.sin(18*x),
       lambda x: -(x+np.sin(x))*np.exp(-x**2),
       lambda x: np.sin(x)+np.sin(10/3*x)+np.log(x)-0.84*x+3,
       lambda x: -sum([k*np.cos((k+1)*x + k) for k in range(1, 6)]),
       lambda x: np.sin(x)+np.sin(2/3*x),
       lambda x: -x*np.sin(x)]
omega = [(-1.5, 11),
        (2.7, 7.5),
        (-10, 10),
        (1.9, 3.9),
        (0, 1.2),
        (-10, 10),
        (2.7, 7.5),
        (-10, 10),
        (3.1, 20.4),
        (0, 10)]
```

Будем рассматривать алгоритм с негладкими минорантами. В качестве параметра надежности возьмем `r=1.1`.
Обозначения : $\textit{Global}$ -- глобальная оценка константы Липшица для производной, $\textit{Local}$ -- локальная оценка

```python
r = 1.1
modes1 = ['Global', 'Local']
modes2 = ['notSmooth', 'Smooth']
```

```python
iterations = np.zeros((10, 4))
mode2 = modes2[0]
for i in range(10):
    plt.figure(figsize=(15, 5))
    for j, mode1 in enumerate(modes1):
        x = np.arange(*omega[i], 0.001)
        y = func[i](x)
        x_real, y_real = x[np.argmin(y)], min(y)
        q, k, xx, yy, node_x, node_y = search_min(func[i], *omega[i], r=r, mode1=mode1, mode2=mode2)
        iterations[i, j] = k
        x_min, y_min, z_ = min(q, key=lambda x: x[1])
        print(f'Функция №{i+1}')
        print(mode1)
        print(f'Количество шагов: {k}')
        print(f'Настоящий минимум: {x_real}, {y_real}')
        print(f'Найденный минимум: {x_min}, {y_min}', '\n')

        ax = plt.subplot(1, len(modes1), j+1)
        plt.plot(x, y)
        plt.plot(xx, yy)    # миноранты
        plt.scatter(node_x, node_y)    # точки измерений на кривой

        ax.set_title(mode1)
    plt.show()
```

А теперь посмотрим на гладкие методы

```python
mode2 = modes2[1]
for i in range(10):
    plt.figure(figsize=(15, 5))
    for j, mode1 in enumerate(modes1):
        x = np.arange(*omega[i], 0.001)
        y = func[i](x)
        x_real, y_real = x[np.argmin(y)], min(y)
        q, k, xx, yy, node_x, node_y = search_min(func[i], *omega[i], r=r, mode1=mode1, mode2=mode2)
        iterations[i, j+2] = k
        x_min, y_min, z_ = min(q, key=lambda x: x[1])
        print(f'Функция №{i+1}')
        print(mode1)
        print(f'Количество шагов: {k}')
        print(f'Настоящий минимум: {x_real}, {y_real}')
        print(f'Найденный минимум: {x_min}, {y_min}', '\n')

        ax = plt.subplot(1, len(modes2), j+1)
        plt.plot(x, y)
        plt.plot(xx, yy)    # миноранты
        plt.scatter(node_x, node_y)    # точки измерений на кривой

        ax.set_title(mode1)
    plt.show()
```

```python
import pandas as pd

iterations = iterations.astype('int')
table = pd.DataFrame({'Not Smooth + Global': iterations[:, 0],
                      'Not Smooth + Local': iterations[:, 1],
                      'Smooth + Global': iterations[:, 2],
                      'Smooth + Local': iterations[:, 3]
                      })
table.index.name = 'функция №'
table
```

В отличие от предыдущих методов, сразу видно, что требуется гораздо меньшее число итераций. Алгоритм работает точно
Видно, что у алгоритмов с гладкими минорантами количество шагов меньше, они работают очень быстро. Так же видно, что следует придавать предпочтение алгоритмам с локальной настройкой
